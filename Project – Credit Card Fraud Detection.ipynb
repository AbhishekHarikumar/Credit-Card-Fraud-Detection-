```python
import pandas as pd 
df = pd.read_csv("creditcard.csv")
```

```python
df.isnull().sum()
```

```python
df["Amount"].describe()
```

```python
genuine = len(df[df.Class == 0])
fraud = len(df[df.Class == 1])
fraud_percentage = (fraud / (genuine + fraud)) * 100
print(genuine)
print(fraud)
print(fraud_percentage)
```

```python
import matplotlib.pyplot as plt 
counts_classes = df.value_counts(df['Class'], sort = True)
counts_classes.plot(kind = "bar", rot =0, color='red')
plt.title("Genuine vs Fradulent")
plt.ylabel("Count")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df['NormalizedAmount'] = scaler.fit_transform(df[['Amount']])
print(df)


```

```python
df.drop(["Amount", "Time"], inplace= True, axis= 1)
```

```python
from sklearn.model_selection import train_test_split
y = df['Class']
X = df.drop('Class', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state=42)

print("Shape of train_X: ", X_train.shape)
print("Shape of test_X: ", X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)

y_pred_dt = decision_tree.predict(X_test)

# Calculating accuracy of Decision Tree
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print("Decision Tree Accuracy:", accuracy_dt)
```

```python
random_forest = RandomForestClassifier(random_state=42)
random_forest.fit(X_train,y_train)

y_pred_rf = random_forest.predict(X_test)

accuracy_rf = accuracy_score(y_test, y_pred_rf)

print(accuracy_rf)
```

```python
from sklearn.metrics import precision_score, confusion_matrix, recall_score, f1_score
def metrics(actuals, predictions):
    print("Accuracy: {:.5f}".format(accuracy_score(actuals, predictions)))
    print("Precision: {:.5f}".format(precision_score(actuals, predictions)))
    print("Recall: {:.5f}".format(recall_score(actuals, predictions)))
    print("F1-score: {:.5f}".format(f1_score(actuals, predictions)))
```

```python
confusion_matrix_dt = confusion_matrix(y_test,y_pred_dt.round())
print("Confusion Matrix - Decision Tree")
print(confusion_matrix_dt)
```

```python
metrics(y_test, y_pred_dt)
```
